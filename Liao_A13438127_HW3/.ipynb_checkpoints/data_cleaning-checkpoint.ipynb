```python
import pandas as pd
import numpy as np
import json
```

```python
data = pd.read_json('./springfield.json')
data
```

```python
# get temperature
temperature = data.iloc[-1, :]['history']
temperature
```

```python
start = temperature['start']
interval = 1800
time_temp = list()
for i in range(len(temperature['data'])):
    time_temp.append([start+interval, temperature['data'][i]])
    start += interval
time_temp
```

```python
with open('time_temp.json', 'w') as outfile:
    json.dump(time_temp, outfile)
```
